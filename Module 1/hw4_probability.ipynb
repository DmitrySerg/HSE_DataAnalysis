### Задача 2
Золотая Лихорадка — страшное, но относительно редкое заболевание: среди тысячи человек им болен в среднем один. Первые 20 лет после заражения болезнь никак не проявляется, но может быть выявлена с помощью специального теста. Тест очень хороший: он ошибается лишь в 5% случаев, то есть если тест сдаёт здоровый человек, то с вероятностью 95% тест покажет, что он здоров, и с вероятностью 5%, что он болен, и наоборот — если тест сдаёт заболевший человек, то тест с вероятностью 95% покажет, что он болен и с вероятностью 5%, что здоров. Случайно выбранный среди всей популяции человек сдаёт тест и — о, ужас! — тест говорит ему, что он болен. Какова вероятность того, что он действительно болен?

**Подсказка.** Эта задача на формулу Байеса. Случайно выбранный человек может быть или болен (событие $H_1$), или здоров (событие $H_2$). Априорные вероятности каждого из этих событий (до сдачи теста) мы знаем. Тест может выдать результат «здоров» или «болен», и вероятности каждого результата зависят от того, здоров человек или болен на самом деле. Наша задача — найти вероятность того, что человек болен, при условии, что тест выдал результат «болен».

$P(H_1) = 0.001$

$P(H_2) = 1-P(H_1) = 0.999$

$P(H_1|"болен_{test}") = \frac{P("болен_{test}"|H_1)\cdot P(H_1)}{P("болен_{test}")} = \frac{0.95 \cdot 0.001}{0.95\cdot 0.001 + 0.999\cdot 0.05}\approx 0.019$

Итак, вероятность того, что человек действительно болен, при условии, что это показал тест, даже не добирает 2%, а значит, в оставшихся 98% случаев, когда человек не болен, но тест показал обратное, мы рискуем назначить ему лечение, спасибо редким событиям

### Задача 5
Пусть $X$ — равномерно распределенная на отрезке $[0, 1]$ случайная величина, а $x_1, …, x_n$ — выборка независимых значений этой случайной величины. Для $n=1, 2, 5, 10, 20, 100$, сгенерировать 1000 выборок, для каждой найти выборочное среднее и построить гистограмму распределения средних значений (для каждого $n$ — свою гистограмму). При построении гистограммы использовать параметр `normed=True`. На тех же графиках построить кривую нормального распределения с математическим ожиданием $0.5$ и стандартным отклонением $\sqrt{\frac{1}{12}}\frac{1}{\sqrt{n}}$. (Здесь $\frac{1}{12}$ — это дисперсия равномерно распределённой случайной величины.) 

**Подсказка.** В `scipy.stats` есть объекты `uniform` и `norm`, соответствующие равномерному и нормальному распределению. Параметры нормального распределения задаются с помощью аргументов `loc`, соответствующего математическому ожиданию, и `scale`, соответствующего (в случае нормального нормального распределения!) стандартному отклонению.

```python
import numpy as np
from __future__ import division
from scipy.stats import uniform, norm
plt.style.use('ggplot')
from pylab import rcParams
rcParams['figure.figsize'] = 6, 6
```

```python
n = [1, 2, 5, 10, 20, 100]
means_dict = {size:[] for size in n}
for n_sample in n:
    sample_means = []
    for i in range(1000):
        sample = uniform.rvs(size=n_sample)
        means_dict[n_sample].append(np.mean(sample))    
```

```python
f, axarr = plt.subplots(3, 2, figsize=(10,15))
index = 0 
for i in range(3):
    for j in range(2):
        axarr[i, j].hist(means_dict[n[index]], bins = 20, normed = True) # больше вложенных индексов богу вложенных индексов
        axarr[i, j].set_title('Sample size {}'.format(n[index]))
        x = np.linspace(min(means_dict[n[index]]), max(means_dict[n[index]]), 200)
        axarr[i, j].plot(x, norm.pdf(x, loc=0.5, scale = np.sqrt(1/12)*1/np.sqrt(n[index])))
        index += 1

f.subplots_adjust(hspace=0.5)
```

Получается красота, распределение выборочных средних всё больше и больше приближается к нормальному с увеличением размера выборки

Выбор очереденого элемента из исходного набора не меняет исходный набор: один и тот же элемент может быть выбран несколько раз (выборка с возвращением). Результат выбора очередного элемента не влияет на выбор следующих элементов. Пусть всего в новой выборке будет $n$ элементов (столько же, сколько исходных). Такая схема часто используется в машинном обучении для приёма, который называется bootstrapping.

1. Какова вероятность того, что конкретный элемент (например, первый) из исходного набора попадёт в новую выборку на первое место (то есть при первом выборе элемента новой выборки)?
2. Какова вероятность того, что конкретный элемент не попадёт в новую выборку на первое место (то есть при первом выборе элемента новой выборки)?
3. Какова вероятность того, что конкретный элемент не попадёт в новую выборку ни на первое, ни на второе место?
3. Какова вероятность того, что конкретный элемент не попадёт в новую выборку ни на первое, ни на второе, ни на третье место?
4. Какова вероятность того, что конкретный элемент вообще не попадёт в новую выборку?
5. Какова вероятность того, что конкретный элемент попадёт в новую выборку?
6. Пусть $X_k$ — случайная величина, принимающая значение 1 в случае, если элемент исходного набора с номером $k$ попал в новую выборку, и значение 0 в противном случае. Найти $EX_k$.
7. Пусть $X$ — случайная величина, равная числу элементов исходной выборки, которые не попали в новую выборку. Найти $EX$. (**Подсказка:** для любых случайных величин $U$ и $V$, матожидание их суммы равно сумме матожиданий: $E(U+V)=EU+EV$.)
9. Пусть $Y$ — случайная величина, равная доле тех элементов исходной выборки, которые не попали в новую выборку. Найти $EY$. (**Подсказка:** Если $U=\lambda V$, то $EU=\lambda EV$.)
10. (\*) Найти предел $EY$ при $n$ стремящемся к бесконечности.

1) Вероятность того, что из первой выборки мы возьмем именно первый элемент, при условии, что мы выбираем их случайно, равна $\frac{1}{n}$, значит, этому же равна и вероятность поставить его на первое место в новенькой выборке

2) Обрратная ситуация - раз мы не можем выбрать конкретный элемент, чтобы поставить его на первое место, то у нас остается еще $n-1$ элементов для расстановки, значит, вероятность такого события - $\frac{n-1}{n} = 1 - \frac{1}{n}$

3) Вероятность того, что мы дважды подряд не выберем один и тот же элемент равна $(1 - \frac{1}{n})^2$

4) По аналогичным рассуждениям - $(1 - \frac{1}{n})^3$

5) Раз мы ни разу не выбираем наш элемент, а всего мы выбираем $n$ раз, то вероятность его не попадания ни на одно из этих мест равна $(1 - \frac{1}{n})^n$
6) Имея уже найденную вероятность того, что элемент не попадет в выборку, найдем вероятность обратного события: $1 -(1 - \frac{1}{n})^n $

7) Зададим функцию плотности распределения нашей случайной величины:

$$
f(X_k) = 
\begin{equation*}
\begin{cases}
1, 1-(1-\frac{1}{n})^n \\
0, (1-\frac{1}{n})^n
\end{cases}
\end{equation*}
$$

Зная функцию плотности, посчитаем матожидание:
$EX_k = 1 \cdot (1-(1-\frac{1}{n})^n) + 0 \cdot (1-\frac{1}{n})^n = 1-(1-\frac{1}{n})^n$

8) Так как матожидание суммы равно сумме матожиданий, то найдем матожидание с.в., которая равна 1, если элемент не попал в новую выборку:

$EX_{k_2} = 1 \cdot (1-\frac{1}{n})^n + 0 \cdot (1-(1-\frac{1}{n})^n) = (1-\frac{1}{n})^n$

Теперь, чтобы узнать матожидание $X$ достаточно просуммировать матожидания:

$EX = \sum_n (1-\frac{1}{n})^n$

9) Раз мы хотим найти долю элементов выборки, не попавших в новую выборку, нужно поделить матожидание числа элементов, не попавших в нее, на общее число элементов $n$:

$EY = \frac{EX}{n} = \frac{1}{n} \sum_n (1-\frac{1}{n})^n$

Промежуточное вычисление значения лимита по правилу Лопиталя переписывать не осилил :-)

$$\lim_{n\to\infty} \frac{1}{n} \sum_n (1-\frac{1}{n})^n = 
\sum_n \lim_{n\to\infty} \frac{1}{n}(1-\frac{1}{n})^n = 
\sum_n \lim_{n\to\infty} \exp [\log(\frac{1}{n}(1-\frac{1}{n})^n)]=\\
\sum_n \exp [\lim_{n\to\infty} (n(\log(1-\frac{1}{n})) - \log(n))] = 
\sum_n \exp [\lim_{n\to\infty} n(\log(1-\frac{1}{n})) - \lim_{n\to\infty} \log(n)] = \\
\sum_n \exp [\lim_{n\to\infty} (\frac{-n}{n-1}) - \lim_{n\to\infty} \log(n)] = 
\sum_n \exp [\lim_{n\to\infty} (\frac{-1}{1-\frac{1}{n}}) - \lim_{n\to\infty} \log(n)]=\\
\sum_n \exp [-1 - \lim_{n\to\infty} \log(n)] = \sum_n \exp [-1 - \infty] = 0$$
