**Определение.** *Обратной матрицей* к квадратной матрице $A$ называется такая матрица $A^{-1}$, что $A^{-1}A=AA^{-1}=E$, где $E$ — единичная матрица (такая матрица, у которой на диагонали стоят единицы, а все остальные элементы нулевые). Обратная матрица соответствует обратному отображению: если взять вектор $y=Ax$, являющийся образом вектора $x$ под действием матрицы $A$, и подействовать на него матрицей $A^{-1}$, то получится снова вектор $x$. Матрица является обратимой (невырожденной) тогда и только тогда, когда её ранг максимален (равен числу строк или столбцов).

**Определение.** *Ортогональным дополнением* $V^\perp$ к линейному подпространству $V \in \mathbb R^n$ называется множество всех векторов, ортогональных к любому векторую из $V$. Иными словами, 

$$V^\perp = \{u\in \mathbb R^n \mid u\perp v \text{ для любого } v \in V\}$$

**Утвреждение.** Сумма размерностей подпростространства $V$ и его ортогонального дополнения равна размерности всего пространства $\mathbb R^n$. Всякий вектор из $u \in \mathbb R^n$ единственным образом раскладывается в сумму $v+v^\perp$, где $v\in V$, $v^\perp \in V^\perp$.

**Определение.** Линией уровня $c$ функции нескольких переменных $y=f(x_1, x_2, \ldots, x_n)$ называется множество точек, в которых функция принимает значение $c$: 

$$\{(x_1, x_2, \ldots, x_n) \in \mathbb R^n\mid f(x_1, x_2, \ldots, x_n)=c\}$$

**Пример.** Для функции $f(x_1, x_2)=x_1^2 + x_2^2$ линиями уровни являются решения уравнения $x_1^2 + x_2^2 = c$, то есть окружности радиуса $\sqrt{c}$.

**Определение.** *Частной производной* функции нескольких переменных $y=f(x_1, x_2, \ldots, x_n)$ по переменной $x_k$, взятой в точке $(x_1^0, x_2^0, \ldots, x_n^0)$ называется значение производной функции одной переменной $x_k$

$$h(x_k) = f(x_1^0, x_2^0, \ldots, x_{k-1}^0, x_k, x_{k+1}^0, \ldots, x_n^0),$$

взятой в точке $x_k^0$:

$$f'_{x_k}(x_1^0, x_2^0, \ldots, x_n^0) = \frac{\partial f}{\partial x_k}(x_1^0, \ldots, x_n^0) = h'(x_k^0)$$

Иными словами, чтобы вычислить частную производную по некоторой переменной, следует зафиксировать все остальные переменные.

*Вектором градиента* функции $f$ называется вектор, составленный из частных производных:

$$\nabla f (x^0)= (f'_{x_1}(x^0), \ldots, f'_{x_n}(x^0))$$

Вектор градиента показывает направление наискорейшего роста функции $f$ вблизи точки $x^0$.

### Задача 11
С помощью метода градиентного спуска попытаться найти минимум у функции $f(x_1, x_2) = x_1 x_2$. Использовать начальное условие $x_1^0 = 1$, $x_2^0 = 1$. Изменить чуть-чуть начальное условие. Как изменится результат? Объяснить поведение алгоритма.

```python
def f(u):
    return u[0] * u[1]
```

```python
def grad(u):
    return np.array([u[1], u[0]])
```

```python
x_old = np.array([1, 1])

def grad_decsent(x_old, delta = 0.1, steps = 100):
    plt.figure(figsize=(6, 6))
    x1, x2 = np.mgrid[-3:3:100j, -3:3:100j] 
    plt.contour(x1, x2, f(np.array([x1, x2])))

    points = []
    for i in range(steps):
        x_new = x_old - delta * grad(x_old)
        x_old = x_new
        points.append(x_new)

    plt.plot([p[0] for p in points], 
             [p[1] for p in points], 'o-', color = 'red')

    x, y = np.mgrid[-3:3:10j, -3:3:10j]
    plt.quiver(x, y, y, x, color='Teal')
    plt.title('Contour levels')
    plt.show()
    
grad_decsent(x_old)
```

Для начальной точки [1, 1] градиентный спуск дошел до седловой точки (0, 0), и остановился в ней

```python
grad_decsent([2, 1], delta=0.01, steps=200)
```

```python
grad_decsent([1, 1.5], delta=0.01, steps=250)
```

Но стоит немного поменять начальное условие, и вот уже алгоритм идет ровненько против возрастания функции и скатывается в одну из областей седла, где функция убывает, и дальше может продолжать свое бессмысленное и беспощадное падение до бесконечности. Красота!
